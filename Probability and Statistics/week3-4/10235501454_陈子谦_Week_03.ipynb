#### 第一组：
![sample1](./sample1.jpg)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson

# 创建二项分布B(5,0.4)概率分布列
x_binom_5 = np.arange(0, 6)  # 横坐标
probabilities_binom_5 = np.array([0.07776, 0.2592, 0.3456, 0.2304, 0.0768, 0.01024])  # 对应概率

# 创建二项分布B(10,0.4)概率分布列
x_binom_10 = np.arange(0, 11)  # 横坐标
probabilities_binom_10 = np.array([0.00605,0.04031,0.12093,0.21499,0.25082,0.20066,0.11148,0.04247,0.01062,0.00157,0.00011])  # 对应概率

# 创建泊松分布 p(4)概率密度函数
x_poisson = np.arange(0, 13)  # 横坐标
λ = 4  # 泊松分布的参数
pmf_poisson = poisson.pmf(x_poisson, λ)  # 计算概率密度函数值

# 绘制概率分布图
plt.figure(figsize=(10, 12))

# 绘制二项分布B(5,0.4)
plt.subplot(3, 1, 1)
plt.bar(x_binom_5, probabilities_binom_5, align='center', color='blue', alpha=0.5)
plt.xlabel('X')
plt.ylabel('Probability')
plt.title('Binomial Distribution B(5,0.4)')
plt.xticks(x_binom_5)
plt.grid(True)

# 绘制二项分布B(10,0.4)
plt.subplot(3, 1, 2)
plt.bar(x_binom_10, probabilities_binom_10, align='center', color='green', alpha=0.5)
plt.xlabel('X')
plt.ylabel('Probability')
plt.title('Binomial Distribution B(10,0.4)')
plt.xticks(x_binom_10)
plt.grid(True)

# 绘制泊松分布 p(4)
plt.subplot(3, 1, 3)
plt.bar(x_poisson, pmf_poisson, align='center', color='red', alpha=0.5)
plt.xlabel('X')
plt.ylabel('Probability')
plt.title('Poisson Distribution (λ=4)')
plt.xticks(x_poisson)
plt.grid(True)

# 绘制累积分布函数图
plt.figure(figsize=(10, 6))

# 二项分布B(5,0.4)的累积分布函数
plt.plot(x_binom_5, binom.cdf(x_binom_5, 5, 0.4), label='Binomial B(5,0.4)', color='blue')

# 二项分布B(10,0.4)的累积分布函数
plt.plot(x_binom_10, binom.cdf(x_binom_10, 10, 0.4), label='Binomial B(10,0.4)', color='green')

# 泊松分布p(4)的累积分布函数
plt.plot(x_poisson, poisson.cdf(x_poisson, 4), label='Poisson p(4)', color='red')

# 添加标签和标题
plt.xlabel('X')
plt.ylabel('CDF')
plt.title('Cumulative Distribution Function')
plt.grid(True)
plt.legend()

# 显示图形
plt.show()
```

### 联系相似与相异：
    相似：
        从他们的概率分布列/概率密度函数能看出二项分布和泊松分布都呈现出一个中间高两边低的离散点形式，他们都是离散型分布。二项分布描述在一系列事件中成功次数的概率，而泊松分布描述在单位时间内事件发生次数的概率，且当二项分布的实验次数n趋向于无穷时，其概率就等于泊松分布的概率。
    相异：
        二项分布的实验次数是有限的，且概率和实验次数有相关性；而泊松分布中只对单位时间的发生次数做统计，不涉及实验次数。两者适用的场景不经相同

#### 第二组：
![sample2](./sample2.jpg)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 定义三个正态分布的参数
params = [(0, 1), (1, 1), (1, 3)]

# 生成 x 轴的值
x = np.linspace(-10, 10, 1000)

# 画图
plt.figure(figsize=(10, 12))
for mu, sigma in params:
    # 计算每个正态分布的概率密度函数
    pdf = norm.pdf(x, mu, sigma)
    # 绘制每个正态分布的概率密度函数图
    plt.subplot(2, 1, 1)
    plt.plot(x, pdf, label=f'μ={mu}, σ={sigma}')

    # 计算每个正态分布的累积分布函数
    cdf = norm.cdf(x, mu, sigma)
    # 绘制每个正态分布的累积分布函数图
    plt.subplot(2, 1, 2)
    plt.plot(x, cdf, label=f'μ={mu}, σ={sigma}')

# 添加标签和标题（概率密度函数）
plt.subplot(2, 1, 1)
plt.title('Probability Density Function of Normal Distributions')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# 添加标签和标题（累积分布函数）
plt.subplot(2, 1, 2)
plt.title('Cumulative Distribution Function of Normal Distributions')
plt.xlabel('X')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```

### 联系相似与相异：
    相似：
        他们作为正态分布都呈现出钟形曲线的形状，且都是连续性分布。
        所有的正态分布都关于其对应的均值x=μ轴对称
    相异：
        在正态分布中μ的取值影响密度函数的对称轴，μ增大，函数向右移，反之亦然；而标准差σ^2的取值则影响密度函数的最大值和分布函数的增长速率，这意味着σ的变化影响着正态分布的尺度发生改变

#### 第三组：
![sample3](./sample3.jpg)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon, gamma

# 定义指数分布的参数
lambda_exp = 1

# 定义第一个伽马分布的参数
alpha1 = 1
beta1 = 0.5

# 定义第二个伽马分布的参数
alpha2 = 0.5
beta2 = 0.5

# 生成 x 轴的值
x = np.linspace(0, 10, 1000)

# 计算指数分布的概率密度函数
pdf_exp = expon.pdf(x, scale=1/lambda_exp)

# 计算第一个伽马分布的概率密度函数
pdf_gamma1 = gamma.pdf(x, a=alpha1, scale=1/beta1)

# 计算第二个伽马分布的概率密度函数
pdf_gamma2 = gamma.pdf(x, a=alpha2, scale=1/beta2)

# 计算指数分布的累积分布函数
cdf_exp = expon.cdf(x, scale=1/lambda_exp)

# 计算第一个伽马分布的累积分布函数
cdf_gamma1 = gamma.cdf(x, a=alpha1, scale=1/beta1)

# 计算第二个伽马分布的累积分布函数
cdf_gamma2 = gamma.cdf(x, a=alpha2, scale=1/beta2)

# 画图
plt.figure(figsize=(10, 12))

# 概率密度函数图
plt.subplot(2, 1, 1)
plt.plot(x, pdf_exp, color='blue', label='Exponential(λ=1)')
plt.plot(x, pdf_gamma1, color='red', label='Gamma (α=1, β=0.5)')
plt.plot(x, pdf_gamma2, color='green', label='Gamma (α=0.5, β=0.5)')
plt.title('Probability Density Functions')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# 累积分布函数图
plt.subplot(2, 1, 2)
plt.plot(x, cdf_exp, color='blue', label='Exponential(λ=1)')
plt.plot(x, cdf_gamma1, color='red', label='Gamma (α=1, β=0.5)')
plt.plot(x, cdf_gamma2, color='green', label='Gamma (α=0.5, β=0.5)')
plt.title('Cumulative Distribution Functions')
plt.xlabel('X')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```

### 联系相似与相异：
    相似：
        他们都是连续性分布，都有参数λ的引入，他们的概率密度函数的右侧都是呈现下降的趋势都描述了在连续空间反胃内随机事件发生的时间和空间间隔
    相异：
        指数分布的概率密度函数都呈现指数型下降的趋势，但伽马分布并不是，题目中所给α和λ都是很特殊的值，当他们都是2的时候就可以从图像中看到呈现先增后减的趋势

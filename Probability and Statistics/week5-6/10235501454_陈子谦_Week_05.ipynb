第二题：对于题目中的图像均在下方绘制出，而对于如何使生成的直方图面积为1，则是在绘制直方图时使用density=True参数。这会将直方图归一化，使得其总面积等于1。根据Python的文档可知，density=True 参数会将直方图的计数转换为概率密度，所以每个条形的面积（条形的高度乘以条形的宽度）表示该区间内值的概率。

```python
import numpy as np
import matplotlib.pyplot as plt
# -*- coding: utf-8 -*-
plt.rcParams['font.family'] = 'HarmonyOS Sans SC'

# 设置随机数生成的数量
n = 10**6

# 生成均匀分布U(0,1)的随机数
uniform_random_numbers = np.random.uniform(0, 1, n)

# 将均匀分布的随机数转换为指数分布Exp(1)的随机数
exponential_random_numbers = -np.log(1 - uniform_random_numbers)

# 绘制指数分布的伪随机数的直方图
plt.hist(exponential_random_numbers, bins=100, density=True, alpha=0.6, color='g')

# 绘制指数分布Exp(1)的密度函数
x = np.linspace(0, np.max(exponential_random_numbers), num=1000)
y = np.exp(-x)
plt.plot(x, y, linewidth=1, color='r')

# 设置图例
plt.legend(['密度函数 Exp(1)', '伪随机数直方图'])

# 显示图表
plt.show()

# 计算 y_i
y_i = 2 * exponential_random_numbers

# 绘制 y_i 的直方图
plt.hist(y_i, bins=100, density=True, alpha=0.6, color='b')

# 设置图例
#plt.legend(['y_i的直方图'])

# 显示图表
#plt.show()

# 绘制指数分布 Exp(0.5) 的密度函数
x = np.linspace(0, np.max(y_i), num=1000)
y = 0.5 * np.exp(-0.5 * x)
plt.plot(x, y, linewidth=2, color='r')

# 设置图例
plt.legend(['密度函数 Exp(0.5)', 'y_i的直方图'])

# 显示图表
plt.show()

# 计算 y_i
y_i = exponential_random_numbers**2

# 绘制 y_i 的直方图
plt.hist(y_i, bins=100, density=True, alpha=0.6, color='b')

# 设置图例
plt.legend(['y_i的直方图'])

# 显示图表
plt.show()
```

### 第一小问：评述直方图和密度函数的关系。
由图中的绿色区域直方图与红色函数（密度函数）我们不难看出，随着样本数量的增加，直方图越来越接近理论上的密度函数，这说明当样本数量越大，样本分布越接近理论分布。
### 第二小问：该直方图对应的分布应为什么？如何验证你的猜想？
分布应为指数分布（从图像可以看出来非常接近指数分布的密度函数），因为当y_i=2x_i时，等同于将指数分布的λ{exp}(-λx)转换为0.5 * 2λ{exp}(-2λx)此时的λ为原来的0.5倍。为了验证这个逻辑，我们可以在一张图中绘制Exp(0.5)与刚刚的直方图，该图例在代码产生的第二张图片蓝色区域已给出
### 附加题：指出该直方图对应的分布应为什么？如何验证你的猜想？
这个直方图对应的分布是一个变换后的分布，它不再是简单的指数分布。由于对每个x_i进行了平方操作，这将导致y_i变得更大。
为此，我们可以考虑的pdf: f(x) = e^{-x}
经过推导我们可以知道p(y) = (1/y^0.5)*{exp}(2y^0.5)
这是一个类似于指数分布的密度函数，但他本身并不是指数分布
